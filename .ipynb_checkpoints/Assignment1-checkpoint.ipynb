```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

url = 'https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/assignment1/Ruter_data.csv'
df = pd.read_csv(url, sep=';')
df
```

```python
df.isna().sum()
```

```python
#Her plotter jeg antalle billetter innenfor de forskjellige kommunene
df["Kommune"].value_counts()
ax = sns.countplot(x="Kommune", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.figure(figsize = (20,10))
```

```python
#Setter alle som har negative passasjerer til 0 og plotter en graf som viser antall passasjerer
for index, row in df.iterrows():
    if(row["Passasjerer_Ombord"]<0):
        df.loc[index, "Passasjerer_Ombord"] = 0

sns.countplot(data=df, x="Passasjerer_Ombord")
```

```python
#Her plotter jeg en tabell over kapasiteten til bussene
sns.countplot(data=df, x="Kjøretøy_Kapasitet")
```

```python
#Her plotter jeg forholdet mellom ledige seter og personer ombord, med en gjennomsnittslinje 
df["Ledige_Seter"] = df["Kjøretøy_Kapasitet"] - df["Passasjerer_Ombord"]
df["Ledige_Seter"].value_counts()
sns.regplot(data=df, x="Passasjerer_Ombord", y="Ledige_Seter");
```

```python
#Her plotter jeg antall biletter det var daglig i datasettet
df['Dato'] = pd.to_datetime(df['Dato'], errors='coerce')
df['Dato'].value_counts().plot.area()
```

```python
#Antall billetter på hver linje
df["Linjenavn"].value_counts().plot.pie()
#Blir veldig uoversiktig mot de lavere tallene
```
