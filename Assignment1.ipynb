```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

url = 'https://raw.githubusercontent.com/umaimehm/Intro_to_AI_2021/main/assignment1/Ruter_data.csv'
df = pd.read_csv(url, sep=';')
```

```python
df["Kommune"].value_counts()
ax = sns.countplot(x="Kommune", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.figure(figsize = (20,10))
```

```python
for index, row in df.iterrows():
    if(row["Passasjerer_Ombord"]<0):
        df.loc[index, "Passasjerer_Ombord"] = 0;
```

```python
df["Ledige_Seter"] = df["Kjøretøy_Kapasitet"] - df["Passasjerer_Ombord"]
df["Ledige_Seter"].value_counts()
ledig = sns.regplot(data=df, x="Passasjerer_Ombord", y="Ledige_Seter")
```

```python
df['Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra'] = pd.to_datetime(df["Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra"], errors='coerce')
df['Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'] = pd.to_datetime(df['Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'], errors='coerce')
df['Tid_mellom_Avgang_og_Ankomst'] = df['Tidspunkt_Faktisk_Avgang_Holdeplass_Fra'] - df['Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra']
df['Tid_mellom_Avgang_og_Ankomst'].value_counts().plot.line(title="Tiden bussen er stillestående ved stasjonen", ylabel="Antall busser", xlabel="Tider")
```
